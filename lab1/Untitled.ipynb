##  9. Определите наибольшее и наименьшее значение годового дохода в категории Genre

## 14. Преобразуйте значения Annual Income в числа с плавающей запятой и получите сводку статистик для столбца Annual Income

## 15. Разделите набор данных на три датафрейма по рейтингу трат (высокий, средний и низкий)

```python
low = df[df["Spending Score (1-100)"] < 35]
mid = df[(df["Spending Score (1-100)"] >= 35) & (df["Spending Score (1-100)"] <= 65)]
high = df[df["Spending Score (1-100)"] > 65]
print(len(low), len(mid), len(high))
```
